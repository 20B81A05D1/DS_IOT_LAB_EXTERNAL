```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
water_quality = pd.read_csv(r"C:\Users\Rutvik\Downloads\Data Science\DS_LAB_Internal -  Question and Datasets - 2023-May\water_quality.csv")
```

```python
water_quality.head()
```

```python
water_quality.isnull().sum()
```

```python
water_quality.ph.fillna(water_quality.ph.median(),inplace=True)
water_quality.Sulfate.fillna(water_quality.Sulfate.median(),inplace=True)
water_quality.Trihalomethanes.fillna(water_quality.Trihalomethanes.median(),inplace=True)
```

```python
water_quality.head()
```

```python
target = water_quality['Quality']
water_quality.drop(['Quality'],axis=1,inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import seaborn as sns
```

```python
x_train,x_test,y_train,y_test = train_test_split(water_quality,target,test_size=0.2,random_state=4)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
```
